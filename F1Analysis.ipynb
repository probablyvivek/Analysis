```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import pandas as pd

# Setup plotting
plotting.setup_mpl()

# Enable the cache
ff1.Cache.enable_cache('/Users/vivektiwari/Library/Caches/')
```

```python
#get the schedule
def get_schedule(year):
    schedule = ff1.get_event_schedule(year)
    schedule = schedule[['RoundNumber', 'Country', 'EventDate', 'EventName', 'EventFormat']]
    return schedule
```

```python
get_schedule(2022)
```

```python
#Get Result of a Race
def get_result(year, race, event):
    session = ff1.get_session(year, race, event)
    session.load()
    session = session.results
    session = session[['FullName', 'Abbreviation', 'TeamName', 'Position','GridPosition', 'Time', 'Status', 'Points']]
    return session
```

```python
#Event is 'R' for Race and 'Q' for qualification
get_result(2022, 'Bahrain', 'R')
```

```python
#Fastest Lap
def fastest_lap(year, race, event):
    session = ff1.get_session(year, race, event)
    session.load()
    session.laps
    fastest_lap = session.laps.pick_fastest()
    time = fastest_lap['LapTime']
    driver = fastest_lap['Driver']
    print(f'Fastest Lap was by {driver} in {time}')
```

```python
fastest_lap(2022, 'Bahrain', 'R')
```
